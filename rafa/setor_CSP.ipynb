```python
import pandas as pd 
```

```python
cep_vs_setor = pd.read_csv('census_code_cep_coordinates.csv', usecols=['POSTCODE', 'CD_GEOCODI'])
ibge_setor =  pd.read_excel('Agregados_preliminares_por_setores_censitarios_SP.xlsx', usecols=['CD_SETOR','NM_MUN'])
ibge_setor.head(5)
```

```python
cep_vs_setor.rename(columns={'CD_GEOCODI': 'CD_SETOR', 'POSTCODE': 'CEP'}, inplace=True)
cep_vs_setor.head(5)
```

```python
dados_sp = ibge_setor[ibge_setor['NM_MUN'] == 'São Paulo']
dados_sp.head(5)
```

```python
dados_sp.loc[:, 'CD_SETOR'] = dados_sp['CD_SETOR'].str.replace('P', '')
dados_sp
```

```python
dados_sp.loc[:, 'CD_SETOR'] = dados_sp['CD_SETOR'].astype(str)
cep_vs_setor.loc[:,'CD_SETOR'] = cep_vs_setor['CD_SETOR'].astype(str)
```

```python
dados_merge = pd.merge(dados_sp, cep_vs_setor, on='CD_SETOR')
```

```python
dados_CEPvsSETOR = dados_merge[['CD_SETOR', 'CEP']]
```

```python
dados_CEPvsSETOR
```

```python

dados_CEPvsSETOR.to_csv('dados_CEPvsSETOR.csv', index=False)
```
