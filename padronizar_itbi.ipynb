```python
import pandas as pd
```

# Arquivo criado para padronizar os arquivos de ITBI.
### Esses arquivos possuem varios meses em Sheets diferentes no Excel, entao esse codigo serve como ferramenta para agrupar todos em um arquivo so e mais simplificado

```python
# Colunas desejadas
columns_to_keep = [
    'N° do Cadastro (SQL)',
    'CEP',
    'Valor Venal de Referência',
    'Valor de Transação (declarado pelo contribuinte)',
    'Proporção Transmitida (%)',
    'Área do Terreno (m2)',
    'Bairro'
]

# Caminho do arquivo Excel
file_paths = ['dados\GUIAS-DE-ITBI-PAGAS-2023.xlsx', 'dados\GUIAS_DE_ITBI_PAGAS_12-2022.xlsx', 'dados\GUIAS_DE_ITBI_PAGAS_(2021).xlsx', 'dados\GUIAS_DE_ITBI_PAGAS_(2020).xlsx']

# Arquivo base filtrado vazio
itbi = pd.DataFrame()
itbi.head()
```

```python
# Processar apenas as abas que correspondem aos meses
for file_path in file_paths:
    # Carregar o arquivo Excel
    excel_data = pd.ExcelFile(file_path)
    for sheet in [s for s in excel_data.sheet_names if '20' in s]:
        try:
            # Ler os dados e converter o número de cadastro para string após a leitura
            sheet_data = pd.read_excel(
                file_path,
                engine='openpyxl',
                sheet_name=sheet,
                usecols=columns_to_keep,
                )
            print(f"Processando a aba {sheet}...")

            # Concatenar os dados filtrados ao DataFrame final
            itbi = pd.concat([itbi, sheet_data], ignore_index=True)
        except Exception as e:
            print(f"Erro ao processar a aba {sheet}: {e}")

    # Garantir que 'N° do Cadastro (SQL)' seja string com 12 dígitos após o concat
    itbi['N° do Cadastro (SQL)'] = itbi['N° do Cadastro (SQL)'].astype(str).str.zfill(11)

    # Verificar o tipo final da coluna e os primeiros valores
    print(itbi['N° do Cadastro (SQL)'].dtype)  # Deve ser string
    print(itbi['N° do Cadastro (SQL)'].head())
```

```python
# Salvar o DataFrame consolidado para um arquivo CSV (opcional)
itbi.to_excel('dados/IBTI_completo.xlsx', index=False)
```
