```python
import pandas as pd
```

# Arquivo criado para padronizar os arquivos de ITBI.
### Esses arquivos possuem varios meses em Sheets diferentes no Excel, entao esse codigo serve como ferramenta para agrupar todos em um arquivo so e mais simplificado

```python
# Colunas desejadas
columns_to_keep = [
    'N° do Cadastro (SQL)',
    'CEP',
    'Valor Venal de Referência',
    'Área do Terreno (m2)',
    'Bairro'
]

# Caminho do arquivo Excel
file_path = 'dados\GUIAS-DE-ITBI-PAGAS-2023.xlsx'

# Arquivo base filtrado vazio
itbi = pd.DataFrame()
itbi.head()
```

```python
# Carregar o arquivo Excel
excel_data = pd.ExcelFile(file_path)
```

```python
# Processar apenas as abas que correspondem aos meses de 2023
for sheet in [s for s in excel_data.sheet_names if '2023' in s]:
    try:
        # Ler apenas as colunas necessárias
        sheet_data = pd.read_excel(file_path, sheet_name=sheet, usecols=columns_to_keep)
        print(f"Processando a aba {sheet}...")
        
        # Concatenar os dados filtrados ao DataFrame final
        itbi = pd.concat([itbi, sheet_data], ignore_index=True)
    except Exception as e:
        print(f"Erro ao processar a aba {sheet}: {e}")
```

```python
# Salvar o DataFrame consolidado para um arquivo CSV (opcional)
itbi.to_excel('dados/IBTI_2023.xlsx', index=False)
```
