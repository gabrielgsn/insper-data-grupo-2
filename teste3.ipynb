```python
# Carregar o shapefile dos setores censitários
setores = gpd.read_file("dados\geo\SP_Malha_Preliminar_2022.shp")

# Calcular o centroide (ponto central) de cada setor censitário
setores['centroid'] = setores.geometry.centroid

# Mostrar os setores com seus centroides (lat, lon)
setores.keys()
```

```python
setores.head()
```

```python
sp = setores[setores["NM_MUN"] == "São Paulo"]
```

```python
sp = sp.to_crs(epsg=4326)
sp.head()
```

```python
from geopy.geocoders import Nominatim

def geocode_address(address):
    try:
        geolocator = Nominatim(user_agent="Creating a conversion table for São Paulo")
        location = geolocator.geocode(address)
        
        if location:
            return (location.latitude, location.longitude)
        else:
            return None
    except:
        return None

# Exemplo de uso:
endereco = "Rua Conde de Irajá 142, São Paulo, Brasil"
coordenadas = geocode_address(endereco)
print(f"Coordenadas do endereço: {coordenadas}")
```

```python
import geopandas as gpd
from shapely.geometry import Point

def find_sector(coordenadas, setores_gdf):
    # Criar um ponto com a lat e long fornecida
    try:
        lat, lon = coordenadas
        point = Point(lon, lat)  # Note que o formato é (longitude, latitude)
        
        # Verificar qual setor contém o ponto
        setor_encontrado = setores_gdf[setores_gdf.contains(point)]
        
        if not setor_encontrado.empty:
            return setor_encontrado["CD_SETOR"].values[0]
        else:
            return None
    except:
        return None

# Exemplo de uso:
setor = find_sector(coordenadas, sp)  # 'setores_gdf' é seu GeoDataFrame com os setores censitários

if setor is not None:
    print(f"O endereço está no setor: {setor}")
else:
    print("Nenhum setor encontrado para o endereço.")
```

```python
import sqlite3
import pandas as pd

# Conectando ao banco de dados SQLite
conn = sqlite3.connect('dados_final.db')

# Consultando os dados da tabela 'dados_final'
query = "SELECT preco_medio_m2, lat, lon FROM dados_final"
df = pd.read_sql_query(query, conn)

# Fechando a conexão
conn.close()

# Visualizar os dados carregados
print(df.head())
```

```python
import folium
from folium.plugins import MarkerCluster

# Criando o mapa centralizado em São Paulo
map_sp = folium.Map(location=[-23.5505, -46.6333], zoom_start=12)

# Adicionando pontos com base no preço/m²
marker_cluster = MarkerCluster().add_to(map_sp)

for idx, row in df.iterrows():
    folium.Marker(
        location=[row['lat'], row['lon']],
        popup=f"Preço/m²: R${row['preco_medio_m2']}",
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(marker_cluster)

# Salvando o mapa em um arquivo HTML
map_sp.save("mapa_preco_m2.html")
```
