```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
from unidecode import unidecode
```

## Testes e informações relevantes

```python
uf = "SP"
cidade = "São Paulo"
endereco = "Rua Topázio"  # Use o nome exato da rua

link = f'https://viacep.com.br/ws/{uf}/{cidade}/{endereco}/json/'

# Realiza a requisição
requisicao = requests.get(link)
dic_requisicao_topazio = requisicao.json()
primeiro_endereco_topazio = dic_requisicao[0]
cep = primeiro_endereco_topazio.get('cep', 'CEP não encontrado')
        
print("CEP encontrado:", cep)
```

```python
for i in range(len(dic_requisicao_topazio)):
    if dic_requisicao_topazio[i].get('logradouro', 'logradouro não encontrado') == endereco:
        cep = dic_requisicao_topazio[i].get('cep', 'CEP não encontrado')
        bairro = dic_requisicao_topazio[i].get('bairro', 'BAIRRO não encontrado')
    print(bairro,':',cep)
```

## Três bairros e Ceps

```python
uf = "sao-paulo"
cidade = "sao-paulo"
bairros = ["vila-mariana","vila-olimpia",'paraisopolis']

bairro_x_cep = {}

for bairro in bairros:

    link = f'https://cepbrasil.org/{uf}/{cidade}/{bairro}/'

    response = requests.get(link)

    if response.status_code == 200:

        html_content = response.text
        soup = BeautifulSoup(html_content, 'html.parser')
        h4_elements = soup.find_all('h4', class_='title')
    
        if bairro not in bairro_x_cep.keys():
            bairro_x_cep[bairro] = []
    
        for h4 in h4_elements:
            texto = (h4.get_text())
            cep = texto[4:12]
            bairro_x_cep[bairro].append(cep)
    else:
        print(f"Erro na requisição: {response.status_code}")

print(bairro_x_cep)
```

```python
df_bairro_x_cep = pd.DataFrame.from_dict(bairro_x_cep, orient='index').transpose()
df_bairro_x_cep
```

## Encontrar todos os bairros de são paulo

```python
bairros_sp_url = 'https://cepbrasil.org/sao-paulo/sao-paulo/'
response_bairro_sp = requests.get(bairros_sp_url)

# Lista para armazenar os bairros
bairros_sp_lista = []

if response_bairro_sp.status_code == 200:
    html_content = response_bairro_sp.text
    soup = BeautifulSoup(html_content, 'html.parser')
    h4_elements = soup.find_all('h4', class_='title')
    
    for h4 in h4_elements:
        bairro = h4.get_text()
        bairros_sp_lista.append(bairro)
else:
    print(f"Erro na requisição: {response_bairro_sp.status_code}")

# Exibir a lista de bairros
print(bairros_sp_lista)
```

```python
bairros_sp_lista_util = []
for bairro in bairros_sp_lista:

    bairro_processado = unidecode(bairro.lower())  
    bairro_processado = bairro_processado.replace(' ', '-')  
    bairro_processado = bairro_processado.replace('(', '-')  
    bairro_processado = bairro_processado.replace(')', '-')  
    
    bairros_sp_lista_util.append(bairro_processado)
```

```python
uf = "sao-paulo"
cidade = "sao-paulo"

bairro_x_cep_total = {}

for bairro in bairros_sp_lista_util:

    link = f'https://cepbrasil.org/{uf}/{cidade}/{bairro}/'

    response = requests.get(link)

    if response.status_code == 200:

        html_content = response.text
        soup = BeautifulSoup(html_content, 'html.parser')
        h4_elements = soup.find_all('h4', class_='title')
    
        if bairro not in bairro_x_cep_total.keys():
            bairro_x_cep_total[bairro] = []
    
        for h4 in h4_elements:
            texto = (h4.get_text())
            cep = texto[4:12]
            bairro_x_cep_total[bairro].append(cep)
            print(bairro)
    else:
        print(f"Erro na requisição: {response.status_code}, {bairro}")

print(bairro_x_cep_total)
```

```python
bairro_x_cep_total = pd.DataFrame.from_dict(bairro_x_cep_total, orient='index').transpose()
bairro_x_cep_total
```
