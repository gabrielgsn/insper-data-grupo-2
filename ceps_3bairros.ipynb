```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

## Testes e informações relevantes

```python
uf = "SP"
cidade = "São Paulo"
endereco = "Rua Topázio"  # Use o nome exato da rua

link = f'https://viacep.com.br/ws/{uf}/{cidade}/{endereco}/json/'

# Realiza a requisição
requisicao = requests.get(link)
dic_requisicao_topazio = requisicao.json()
primeiro_endereco_topazio = dic_requisicao[0]
cep = primeiro_endereco_topazio.get('cep', 'CEP não encontrado')
        
print("CEP encontrado:", cep)
```

```python
for i in range(len(dic_requisicao_topazio)):
    if dic_requisicao_topazio[i].get('logradouro', 'logradouro não encontrado') == endereco:
        cep = dic_requisicao_topazio[i].get('cep', 'CEP não encontrado')
        bairro = dic_requisicao_topazio[i].get('bairro', 'BAIRRO não encontrado')
    print(bairro,':',cep)
```

## Três bairros e Ceps

```python
uf = "sao-paulo"
cidade = "sao-paulo"
bairros = ["vila-mariana","vila-olimpia",'paraisopolis']

bairro_x_cep = {}

for bairro in bairros:

    link = f'https://cepbrasil.org/{uf}/{cidade}/{bairro}/'

    

    response = requests.get(link)

    if response.status_code == 200:

        html_content = response.text
        soup = BeautifulSoup(html_content, 'html.parser')
        h4_elements = soup.find_all('h4', class_='title')
    
        if bairro not in bairro_x_cep.keys():
            bairro_x_cep[bairro] = []
    
        for h4 in h4_elements:
            texto = (h4.get_text())
            cep = texto[4:12]
            bairro_x_cep[bairro].append(cep)
    else:
        print(f"Erro na requisição: {response.status_code}")

print(bairro_x_cep)
```

```python
df_bairro_x_cep = pd.DataFrame.from_dict(bairro_x_cep, orient='index').transpose()
df_bairro_x_cep
```
